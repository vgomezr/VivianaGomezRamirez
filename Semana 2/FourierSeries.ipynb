```python
%matplotlib notebook 
import numpy as np
import matplotlib.pyplot as plt
from celluloid import Camera
import scipy as sp
from scipy import integrate
```

```python
def f(x,n):
    f = (2/n) * (-1)**(n-1) * np.sin(n*x)
    return f
```

```python
L = np.pi
cicles = 3
x = np.linspace(0,2*cicles*L,100)
```

```python
n = 1
nt = 50
F = 0
```

```python
fig = plt.figure(figsize=(6,6))
camera = Camera(fig)
```

```python
while n < nt:
    F += f(x,n)
    plt.plot(x,F,c='k')
    camera.snap()
    n += 1 
```

```python
animation = camera.animate()
animation.save('Fourier.mp4')
```
