$\textbf{Derivada central de la función}$ $f(x)=e^{-x^{2}}$

```python
file = 'DatosDerivada.dat.txt' #Datos obtenidos con C++
data= np.loadtxt(file)
```

```python
plt.plot(data[:,0], data[:,1])
plt.xlabel("x")
plt.ylabel("f´(x)")
```

```python
x= np.linspace(-20,20,100)
f= lambda x: -2*(np.e**(-x**2))*x
analitica= f(x)
plt.plot(x,analitica,  "--", c="r",label= "Derivada analítica")
plt.legend()
```
