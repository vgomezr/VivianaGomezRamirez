```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from scipy.integrate import solve_ivp
from scipy import integrate
from scipy.optimize import fsolve
```

## Ecuación diferencial no lineal

$${\frac{du}{dt}} = u^q$$

```python
t= np.linspace(0,10,30)
```

```python
#Con la solución analítica, es claro que u(0)=1.
t0= 0
u0=1
r0=[t0,u0]
```

```python
q=[0., 0.2, 0.4, 0.7, 0.9, 1.]
```

```python
soluciones=[]
def solve():
    for i in range(len(q)):
        q_i= q[i]
        def df(u,q):
            return u**(q_i)
        
        sol= solve_ivp(df, (t[0], t[-1]), r0, t_eval= t, method= "RK45")
        plt.plot(sol.t, sol.y[0], label= r"$q={}$".format(q_i))
        plt.legend()
        plt.show()
```

```python
solve()
```
